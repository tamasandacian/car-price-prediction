# Data Wrangling / Data Cleansing / Data Cleaning

```python
# Data Wrangling is the process of converting data from the initial format to a format that may be better for analysis.
import pandas as pd
import matplotlib.pylab as plt
```

```python
# Reading the dataset from URL and add the related headers
filename = "https://s3-api.us-geo.objectstorage.softlayer.net/cf-courses-data/CognitiveClass/DA0101EN/auto.csv"
headers = ["symboling","normalized-losses","make","fuel-type","aspiration", "num-of-doors","body-style",
         "drive-wheels","engine-location","wheel-base", "length","width","height","curb-weight","engine-type",
         "num-of-cylinders", "engine-size","fuel-system","bore","stroke","compression-ratio","horsepower",
         "peak-rpm","city-mpg","highway-mpg","price"]
```

```python
df = pd.read_csv(filename, names = headers)
```

```python
# To see what the data set looks like, we'll use head() method
df.head()
```

# Identify and handle missing values

```python
import numpy as np
#(1) Convert "?" to NaN
df.replace("?", np.nan, inplace=True)
df.head()
```

# Evaluating for Missing Data

```python

# There are two methods for detecting in Python the missing values 
# (1) .isnull() method
# (2) .notnull() method
missing_data = df.isnull()
missing_data.head()
```

```python
# True stands for missing data, False stands for not missing data
```

```python
# Count missing values in each column
for column in missing_data.columns.values.tolist():
    print(column)
    print(missing_data[column].value_counts())
    print("")
```

```python
# Summary:
# normalized-losses: 41 missing data
# num-of-doors: 2 missing data
# bore: 4 missing data
# stroke: 4 missing data
# horsepower: 2 missing data
# peak-rpm: 2 missing data
# price: 4 missing data
```

```python
# Dealing with missing data
# (1) drop data
# a. drop the whole row
# b. drop the whole column
# (2) replace data
# a. replace it by mean
# b. replace it by frequency
# c. replace it based on other functions
```

# Replace by mean

```python
# normalized-losses: 41 missing data, replace it by mean
# bore: 4 missing data, replace it by mean
# stroke: 4 missing data, replace it by mean
# horsepower: 2 missing data, replace it by mean
# peak-rpm: 2 missing data, replace it by mean
```

```python
# Calculate the average of the column "normalized-losses"
avg_norm_loss = df["normalized-losses"].astype("float").mean(axis=0)
print("Average of normalized-losses:", avg_norm_loss)
```

```python
# Replace NaN with the mean value in "normalized-losses" column
df["normalized-losses"].replace(np.nan, avg_norm_loss, inplace=True)
```

```python
# Calculate the mean value for 'bore' column
avg_bore = df["bore"].astype("float").mean(axis=0)
print("Average of bore:", avg_bore)
```

```python
# Replace NaN with the mean value in "bore" column
df["bore"].replace(np.nan, avg_bore, inplace=True)
```

```python
# Calculate the mean value of the "stroke" column
avg_stroke = df["stroke"].astype("float").mean(axis=0)
print("Average of stroke:", avg_stroke)
```

```python
# Replace NaN with the mean value for stroke
df["stroke"].replace(np.nan, avg_stroke, inplace=True)
```

```python
# Calculate the mean value of the "horsepower" column
avg_horsepower = df["horsepower"].astype("float").mean(axis=0)
print("Average of horsepower:", avg_horsepower)
```

```python
# Replace NaN with the mean value on "horsepower" column
df["horsepower"].replace(np.nan, avg_horsepower, inplace=True)
```

```python
# Calculate the mean of the column "peak-rpm"
avg_peak_rpm = df["peak-rpm"].astype("float").mean(axis=0)
print("Average of peak-rpm:", avg_peak_rpm)
```

```python
# Replace NaN with the mean value on "peak-rpm" column
df["peak-rpm"].replace(np.nan, avg_peak_rpm, inplace=True)
```

# Replace by frequency

```python
# num-of-doors: 2 missing data, replace them with "four"
# The reason is that 84% sedans is four doors. Since four doors is most frequent, it is most likely to occur.
df["num-of-doors"].value_counts()
```

```python
# we can determine the most common cars with the num-of-doors
df["num-of-doors"].value_counts().idxmax()
```

```python
# replace the missing values in "num-of-doors" with the most common type - "four"
df["num-of-doors"].replace(np.nan, "four", inplace=True)
```

# Drop the whole row

```python
# price: 4 missing data, simply delete the whole row
# The reason is that price is what we wat to predict. Any data entry without price data cannot be used for prediction.
# Any row without price data is not useful to us.

# we simply drop whole row with NaN in "price" column
df.dropna(subset=["price"], axis=0, inplace=True)

# we reset the index, because we dropped two rows
df.reset_index(drop=True, inplace=True)
```

```python
# we check for the last change
df.head(10)
```

# Data Standardization

```python
# Data Standardization is a term for of data normalization, where we substract the mean and divide by the standar deviation
```

```python
# Example: we transform mpg to L/100km
# formula L/100km = 235 / mpg

# Convert mpg to L/100km by mathematical operation (235 divided by mpg)
df["city-L/100km"] = 235/df["city-mpg"]

# check your transformed data
df.head()
```

```python
# transform mpg to L/100km in the column "highway-mpg" and change the name of the column to "highway-L/100km"

# trasform mpg to L/100km using mathematical operation (235 divided by mpg)
df["highway-mpg"] = 235/df["highway-mpg"]

# rename the column "highway-mpg" to "highway-L/100km"
df.rename(columns={"highway-mpg": "highway-L/100km"}, inplace=True)

# check the transformed data
df.head()
```

# Data Normalization

```python
# Normalization is the process of transforming values of several variables into a similar range.
# Normalization include scaling the variable so the variable values range from 0 to 1
```

```python
# we want to normalize columns: "length", "width", "height"
# the target is to normalize variables so their values ranges from 0 to 1
```

```python
# replace (original value) by (original value)/(maximum value)
# Simple Feature Scaling
df["length"] = df["length"]/df["length"].max()
df["width"] = df["width"]/df["width"].max()
df["height"] = df["height"]/df["height"].max()

# check the transformed data
df[["length", "width", "height"]].head()
```

```python
# Here we can see, we have normalized "length", "width" and "height" in the range of [0,1]
```

# Binning

```python
# Binning is the process of transforming continuous numerical variables into categorical bins, for grouped analysis
```

```python
# Example: "horsepower" is a variable ranging from 48 to 288 (it has 57 unique values)
# we can categorize the price difference with high horsepower, medium horsepower, small horsepower (3 types) for analysis
```

```python
df["horsepower"].dtypes
```

```python
# convert "horsepower" type from object to int type
df["horsepower"] = df["horsepower"].astype(int, copy=True)
```

```python
df["horsepower"].dtypes
```

```python
# we can plot the histogram of horsepower to see what the distribution of horsepower looks like
%matplotlib inline
import matplotlib as plt
from matplotlib import pyplot
plt.pyplot.hist(df["horsepower"])

# set x/y labels and plot title
plt.pyplot.xlabel("horsepower")
plt.pyplot.ylabel("count")
plt.pyplot.title("horsepower bins")
```

```python
# we can plot 3 bins of equal size
bins = np.linspace(min(df["horsepower"]), max(df["horsepower"]), 4)
bins
```

```python
# we set the group names
group_names = ['Low', 'Medium', 'High']
```

```python
group_names
```

```python
# we apply function cut to determine what each value of df["horsepower"] belongs to
df['horsepower-binned'] = pd.cut(df['horsepower'], bins, labels=group_names, include_lowest=True)
df[['horsepower', 'horsepower-binned']].head(20)
```

```python
# check number of vehicles in each bin
df["horsepower-binned"].value_counts()
```

```python
# plot distribution of each bin
%matplotlib inline
import matplotlib as plt
from matplotlib import pyplot
pyplot.bar(group_names, df["horsepower-binned"].value_counts())

# set x/y labels and plot title
plt.pyplot.xlabel("horsepower")
plt.pyplot.ylabel("count")
plt.pyplot.title("horsepower bins")
```

# Indicator Variable (or dummy variable)

```python
# An indicator variable (or dummy variable) is a numerical variable used to label categories.
# They are called 'dummies' because the numbers themselves don't have inherent meaning
```

```python
# It is used for displaying categorical variables in regression analysis
```

```python
# e.g. column "fuel-type" has: "gas" and "diesel" (we convert fuel-type to indicator variables)
df.columns
```

```python
dummy_variable_1 = pd.get_dummies(df["fuel-type"])
```

```python
dummy_variable_1.head()
```

```python
# change column names for clarity
dummy_variable_1.rename(columns={'fuel-type-diesel': 'gas', 'fuel-type-diesel': 'diesel'}, inplace=True)
```

```python
dummy_variable_1.head()
```

```python
# merge data frame "df" and "dummy_variable_1"
df = pd.concat([df, dummy_variable_1], axis=1)

# drop original column "fuel-type" from "df"
df.drop("fuel-type", axis=1, inplace=True)
```

```python
# check transformed data
df.head()
```

```python
# create indicator variable to the column of "aspiration": "std" to 0, while "turbo" to 1

# create dummy_variable for "aspiration" column
dummy_variable_2 = pd.get_dummies(df['aspiration'])

# change the name of newly created columns for clarity
dummy_variable_2.rename(columns={'std': 'aspiration-std', 'turbo': 'aspiration-turbo'}, inplace=True)

# check the first 5 instances of dummy_variable_2
dummy_variable_2.head()
```

```python
# merge the new dataframe to the original dataframe and then drop the column "aspiration"

# merge dummy_variable_2 to the original dataframe
df = pd.concat([df, dummy_variable_2], axis=1)

# drop original column "aspiration"
df.drop("aspiration", axis=1, inplace=True)
```
